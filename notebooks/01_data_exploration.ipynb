```python
import fastf1 as ff1
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
import fastf1.plotting

cache_dir = 'f1_cache'

if not os.path.exists(cache_dir):
    os.makedirs(cache_dir)

ff1.Cache.enable_cache(cache_dir) 

sns.set_style('darkgrid')

print("Cache enabled successfully!")
```

```python
session = ff1.get_session(2023, 'Bahrain', 'R')
session.load()
```

```python
laps = session.laps
```

```python
laps.head()
```

```python
print("First 10 rows of the laps DataFrame:")
display(laps.head(10))

print("\nDataFrame Info:")
laps.info()
```

```python
laps['LapTimeSeconds'] = laps['LapTime'].dt.total_seconds()

fastest_lap_time = laps['LapTimeSeconds'].min()
laps = laps[laps['LapTimeSeconds'] < fastest_lap_time * 1.1]

fastest_laps_per_driver = laps.loc[laps.groupby('Driver')['LapTimeSeconds'].idxmin()]

fastest_laps_summary = fastest_laps_per_driver[['Driver', 'LapNumber', 'LapTimeSeconds']].sort_values(by='LapTimeSeconds').reset_index(drop=True)

print("Fastest Race Lap by Each Driver:")
display(fastest_laps_summary)
```

```python
TEAM_COLORS_MANUAL = {
    'Red Bull Racing': '#0600EF',
    'Ferrari': '#DC0000',
    'Mercedes': '#00D2BE',
    'Alpine': '#0090FF',
    'McLaren': '#FF8700',
    'Alfa Romeo': '#900000',
    'Aston Martin': '#006F62',
    'Haas F1 Team': '#FFFFFF',
    'AlphaTauri': '#2B4562',
    'Williams': '#005AFF'
}

fig, ax = plt.subplots(figsize=(15, 8))

drivers_to_plot = ['VER','ALO','HAM']

for driver in drivers_to_plot:
    driver_laps = laps.pick_driver(driver)
    
    team_name = driver_laps['Team'].iloc[0]

    team_color = TEAM_COLORS_MANUAL.get(team_name, '#000000') 

    ax.plot(
        driver_laps['LapNumber'], 
        driver_laps['LapTimeSeconds'], 
        color=team_color,
        label=driver
    )


ax.set_title(f"{session.event['EventName']} {session.event.year} - Race Pace Comparison", fontsize=16)
ax.set_xlabel("Lap Number", fontsize=12)
ax.set_ylabel("Lap Time (seconds)", fontsize=12)
ax.legend()
ax.grid(True, which='both', linestyle='--', linewidth=0.5)


plt.show()
```

```python
print("Tire compounds used in the race:")
print(laps['Compound'].unique())

laps_with_stint = laps[['Driver', 'Stint', 'Compound', 'LapTimeSeconds']]
fastest_by_compound = laps_with_stint.loc[laps_with_stint.groupby(['Driver', 'Compound'])['LapTimeSeconds'].idxmin()]
fastest_by_compound = fastest_by_compound.sort_values(by='LapTimeSeconds').reset_index(drop=True)

print("\nFastest Lap on Each Compound for Each Driver:")
display(fastest_by_compound)
```

```python
drivers_to_analyze = ['VER', 'PER', 'ALO', 'SAI']
laps_subset = laps[laps['Driver'].isin(drivers_to_analyze)]

plt.figure(figsize=(15, 8))
sns.boxplot(data=laps_subset, x="Driver", y="LapTimeSeconds", hue="Compound", 
            order=drivers_to_analyze,
            hue_order=['SOFT', 'MEDIUM', 'HARD']) 

plt.title("Lap Time Distribution by Driver and Tire Compound")
plt.xlabel("Driver")
plt.ylabel("Lap Time (seconds)")
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.show()
```

```python
laps_race_pace = laps.loc[laps['IsAccurate'] == True]

consistency_metric = laps_race_pace.groupby('Driver')['LapTimeSeconds'].std().sort_values()

print("Driver Race Pace Consistency (Lower is Better):")
print(consistency_metric)

plt.figure(figsize=(12, 6))
consistency_metric.plot(kind='bar')
plt.title("Driver Consistency During 2023 Bahrain GP")
plt.xlabel("Driver")
plt.ylabel("Standard Deviation of Lap Time (s)")
plt.grid(axis='y', linestyle='--', linewidth=0.5)
plt.show()
```

```python
from tqdm import tqdm

ff1.Cache.enable_cache('f1_cache')

schedule_2023 = ff1.get_event_schedule(2023)

past_events = schedule_2023[schedule_2023['EventDate'] < pd.to_datetime('today')]

print("Races to be analyzed:")
display(past_events[['EventName', 'EventDate']])

all_races_consistency = []

for index, event in tqdm(past_events.iterrows(), total=past_events.shape[0], desc="Analyzing Races"):
    try:
        session = ff1.get_session(event['EventDate'].year, event['RoundNumber'], 'R')
        session.load()
        
        laps = session.laps
        laps['LapTimeSeconds'] = laps['LapTime'].dt.total_seconds()
        
        laps_race_pace = laps[laps['IsAccurate'] == True]
        
        race_consistency = laps_race_pace.groupby('Driver')['LapTimeSeconds'].std().reset_index()
        race_consistency.rename(columns={'LapTimeSeconds': 'ConsistencyStd'}, inplace=True)
        
        race_consistency['RaceName'] = event['EventName']
        race_consistency['Year'] = event['EventDate'].year
        
        all_races_consistency.append(race_consistency)
        
    except Exception as e:
        print(f"Could not analyze {event['EventName']}: {e}")

season_consistency_df = pd.concat(all_races_consistency, ignore_index=True)

print("\n\nSample of Season-Wide Consistency Data:")
display(season_consistency_df.head())
```

```python
average_season_consistency = season_consistency_df.groupby('Driver')['ConsistencyStd'].mean().sort_values().dropna()

print("Average Driver Consistency for the 2023 Season (Lower is Better):")
print(average_season_consistency)

plt.figure(figsize=(15, 8))
average_season_consistency.plot(kind='bar')
plt.title("Average Driver Consistency - 2023 Season")
plt.xlabel("Driver")
plt.ylabel("Mean Standard Deviation of Lap Time (s)")
plt.grid(axis='y', linestyle='--', linewidth=0.5)
plt.show()
```

```python
import numpy as np


all_driver_results = []

for index, event in tqdm(past_events.iterrows(), total=past_events.shape[0], desc="Fetching Season Results"):
    
    try:
        session = ff1.get_session(event['Session5Date'].year, event['EventName'], 'R')
        session.load(laps=False, telemetry=False, weather=False, messages=False)
        results = session.results
        
        for _, driver in results.iterrows():
            all_driver_results.append({
                'RaceName': event['EventName'],
                'Driver': driver['Abbreviation'],
                'Team': driver['TeamName'], 
                'FinishingPosition': driver['Position'],
                'GridPosition': driver['GridPosition'],
                'Points': driver['Points'],
                'Status': driver['Status']
            })
    except Exception as e:
        print(f"Could not load Grand Prix results for {event['EventName']}: {e}")
        
    try:
        sprint_session = ff1.get_session(event['Session5Date'].year, event['EventName'], 'Sprint')
        sprint_session.load(laps=False, telemetry=False, weather=False, messages=False)
        sprint_results = sprint_session.results
        
        for _, driver in sprint_results.iterrows():
             all_driver_results.append({
                'RaceName': f"{event['EventName']} Sprint",
                'Driver': driver['Abbreviation'],
                'Team': driver['TeamName'], 
                'FinishingPosition': np.nan,
                'GridPosition': np.nan,
                'Points': driver['Points'],
                'Status': 'Finished'
            })
    except Exception as e:
        pass

season_results_df = pd.DataFrame(all_driver_results)


season_results_df['FinishingPosition'] = pd.to_numeric(season_results_df['FinishingPosition'], errors='coerce')
season_results_df['GridPosition'] = pd.to_numeric(season_results_df['GridPosition'], errors='coerce')
```

```python
driver_performance_metrics = season_results_df.groupby('Driver').agg(
    Team=('Team', 'last'), 
    AvgFinishingPosition=('FinishingPosition', 'mean'),
    AvgGridPosition=('GridPosition', 'mean'),
    TotalPoints=('Points', 'sum'),
    DNFs=('Status', lambda x: (x != 'Finished').sum())
).reset_index()

avg_consistency = season_consistency_df.groupby('Driver')['ConsistencyStd'].mean().reset_index()

final_driver_dataset = pd.merge(driver_performance_metrics, avg_consistency, on='Driver', how='inner')

final_driver_dataset.set_index('Driver', inplace=True)

print("\n\nFinal Combined Dataset for Modeling (with Team Name):")
display(final_driver_dataset.head())
```

```python
final_driver_dataset = final_driver_dataset.sort_values(by='TotalPoints', ascending=False)

final_driver_dataset['ChampionshipRank'] = range(1, len(final_driver_dataset) + 1)

print("\n\nDataset with Target Variable (ChampionshipRank):")
display(final_driver_dataset)
```

```python
output_directory = '../data/processed'
output_filename = 'f1_2023_driver_performance.csv'
output_path = os.path.join(output_directory, output_filename)

os.makedirs(output_directory, exist_ok=True)

final_driver_dataset.to_csv(output_path, index=True)

print(f"Dataset successfully exported to the correct location: {output_path}")
```
