```python
import fastf1 as ff1
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
import fastf1.plotting

cache_dir = 'f1_cache'

if not os.path.exists(cache_dir):
    os.makedirs(cache_dir)

ff1.Cache.enable_cache(cache_dir) 

sns.set_style('darkgrid')

print("Cache enabled successfully!")
```

```python
session = ff1.get_session(2023, 'Bahrain', 'R')
session.load()
```

```python
laps = session.laps
```

```python
laps.head()
```

```python
print("First 10 rows of the laps DataFrame:")
display(laps.head(10))

print("\nDataFrame Info:")
laps.info()
```

```python
laps['LapTimeSeconds'] = laps['LapTime'].dt.total_seconds()

fastest_lap_time = laps['LapTimeSeconds'].min()
laps = laps[laps['LapTimeSeconds'] < fastest_lap_time * 1.1]

fastest_laps_per_driver = laps.loc[laps.groupby('Driver')['LapTimeSeconds'].idxmin()]

fastest_laps_summary = fastest_laps_per_driver[['Driver', 'LapNumber', 'LapTimeSeconds']].sort_values(by='LapTimeSeconds').reset_index(drop=True)

print("Fastest Race Lap by Each Driver:")
display(fastest_laps_summary)
```

```python
TEAM_COLORS_MANUAL = {
    'Red Bull Racing': '#0600EF',
    'Ferrari': '#DC0000',
    'Mercedes': '#00D2BE',
    'Alpine': '#0090FF',
    'McLaren': '#FF8700',
    'Alfa Romeo': '#900000',
    'Aston Martin': '#006F62',
    'Haas F1 Team': '#FFFFFF',
    'AlphaTauri': '#2B4562',
    'Williams': '#005AFF'
}

fig, ax = plt.subplots(figsize=(15, 8))

drivers_to_plot = ['VER','ALO']

for driver in drivers_to_plot:
    driver_laps = laps.pick_driver(driver)
    
    team_name = driver_laps['Team'].iloc[0]

    team_color = TEAM_COLORS_MANUAL.get(team_name, '#000000') 

    ax.plot(
        driver_laps['LapNumber'], 
        driver_laps['LapTimeSeconds'], 
        color=team_color,
        label=driver
    )


ax.set_title(f"{session.event['EventName']} {session.event.year} - Race Pace Comparison", fontsize=16)
ax.set_xlabel("Lap Number", fontsize=12)
ax.set_ylabel("Lap Time (seconds)", fontsize=12)
ax.legend()
ax.grid(True, which='both', linestyle='--', linewidth=0.5)


plt.show()
```

```python
print("Tire compounds used in the race:")
print(laps['Compound'].unique())

laps_with_stint = laps[['Driver', 'Stint', 'Compound', 'LapTimeSeconds']]
fastest_by_compound = laps_with_stint.loc[laps_with_stint.groupby(['Driver', 'Compound'])['LapTimeSeconds'].idxmin()]
fastest_by_compound = fastest_by_compound.sort_values(by='LapTimeSeconds').reset_index(drop=True)

print("\nFastest Lap on Each Compound for Each Driver:")
display(fastest_by_compound)
```

```python
drivers_to_analyze = ['VER', 'PER', 'ALO', 'SAI']
laps_subset = laps[laps['Driver'].isin(drivers_to_analyze)]

plt.figure(figsize=(15, 8))
sns.boxplot(data=laps_subset, x="Driver", y="LapTimeSeconds", hue="Compound", 
            order=drivers_to_analyze,
            hue_order=['SOFT', 'MEDIUM', 'HARD']) 

plt.title("Lap Time Distribution by Driver and Tire Compound")
plt.xlabel("Driver")
plt.ylabel("Lap Time (seconds)")
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.show()
```

```python
laps_race_pace = laps.loc[laps['IsAccurate'] == True]

consistency_metric = laps_race_pace.groupby('Driver')['LapTimeSeconds'].std().sort_values()

print("Driver Race Pace Consistency (Lower is Better):")
print(consistency_metric)

plt.figure(figsize=(12, 6))
consistency_metric.plot(kind='bar')
plt.title("Driver Consistency During 2023 Bahrain GP")
plt.xlabel("Driver")
plt.ylabel("Standard Deviation of Lap Time (s)")
plt.grid(axis='y', linestyle='--', linewidth=0.5)
plt.show()
```
